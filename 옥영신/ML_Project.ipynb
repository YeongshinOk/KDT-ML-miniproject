```python
# 결측값이 94%인 agent, company 제거
hotel = hotel.drop('agent', axis=1)
hotel = hotel.drop('company', axis=1)
```

```python
# 이름 간소화
hotel.columns = ['Hotel', 'Canceled', 'LeadTime', 'ArrivingYear', 'ArrivingMonth', 'ArrivingWeek','ArrivingDate', 'WeekendStay',
              'WeekStay', 'Adults', 'Children', 'Babies', 'Meal','Country', 'Segment', 'DistChannel','RepeatGuest', 'PrevCancel',
              'PrevBook', 'BookRoomType','AssignRoomType', 'ChangeBooking', 'DepositType', 'WaitingDays', 
              'CustomerType', 'ADR','ParkSpace', 'SpecialRequest','Reservation', 'ReservationDate']
```

```python
# 연속형과 범주형 데이터 확인 함수
def var(hotel):
    unique_list = pd.DataFrame([[i,len(hotel[i].unique())] for i in hotel.columns])
    unique_list.columns = ['name','uniques']

    total_var = set(hotel.columns)
    cat_var = set(unique_list.name[(unique_list.uniques<=12)      | 
                                   (unique_list.name=='Country')  | 
                                   (unique_list.name=='Agent')                                     
                                  ])
    con_var = total_var - cat_var
    
    return cat_var, con_var 


cat_var, con_var = var(hotel)

print("연속형 변수 (",len(con_var),")\n",con_var,'\n\n'
      "범주형 변수(",len(cat_var),")\n",cat_var)
```

```python
# Boxplot으로 연속형 변수 이상치 확인

features = ['LeadTime', 'PrevCancel', 'ADR', 'WeekStay', 'Adults', 'WaitingDays', 'ArrivingWeek', 'PrevBook', 'ReservationDate', 'ArrivingDate', 'ChangeBooking', 'WeekendStay']

plt.figure(figsize=(16, 18))
n = 1
for feature in features:
    plt.subplot(4,4,n)
    sns.boxplot(x=hotel[feature])
    plt.title(feature)
    plt.xlabel(feature)
    n+=1
    plt.tight_layout()
```

```python
# 연속형 변수 이상치 처리

# 예약날짜와 체크인 날짜 간의 기간 500일 넘으면 500일로 처리 - 그 이상은 너무 길어서 머신러닝에 악영향을 줄 듯
hotel.loc[hotel.LeadTime >  500, 'LeadTime']=500

# 이전 취소 횟수 0 초과 시 모두 1로 처리 - 취소 전적 유무만 확인
hotel.loc[hotel.PrevCancel >  0, 'PrevCancel']=1

# 숙박일 10일 초과 시 10으로 처리 - 장기투숙은 not common so
hotel.loc[hotel.WeekStay >  10, 'WeekStay']=10

#  성인 수 4명 초과 시 4로 처리 - 그 이상은 데이터가 이상한 것 같음. 단체투숙객 데이터가 잘못 들어간 듯
hotel.loc[hotel.Adults > 4, 'Adults']=  4

# 예약처리 대기일 수 0 초과 시 1로 처리 - 대기일 유무만 확인
hotel.loc[hotel.WaitingDays >  0, 'WaitingDays']=1

# 고객이 이전에 투숙한 횟수인데 0, 1로 유무만 확인하겠음.
hotel.loc[hotel.PrevBook >  0, 'PrevBook']=1

# 예약 변경 횟수 5회 넘어가면 그것 또한 평범하지 않으니 그 이상은 5로 처리
hotel.loc[hotel.ChangeBooking >  5, 'ChangeBooking']=5

# 주말 숙박만 5일 이상 넘어가면 그것도 장기 투숙이니 일반적이지 않으므로 5일 이상은 5로 고정
hotel.loc[hotel.WeekendStay >=  5, 'WeekendStay']=5

cat_var=list(cat_var)
cat_var = set(cat_var + ['PrevBook', 'PrevCancel'])

con_var = set(hotel.columns) - cat_var
con_var=list(con_var)

hotel[con_var].describe()
```

```python
# 카테고리형 변수 이상치 확인
cat_var=list(cat_var)
hotel[cat_var].describe()
```

```python
# 아기 8명 초과는 이상치로 간주 0으로 치환
hotel.loc[hotel.Babies > 8,'Babies'] = 0

# 요청한 주차공간 5개 이상도 이상치로 간주 0으로 치환
hotel.loc[hotel.ParkSpace > 5,'ParkSpace'] = 0

# 아이 수 8명 초과도 이상치로 간주 0으로 치환
hotel.loc[hotel.Children  > 8,'Children']  = 0

hotel[cat_var].describe()
```

```python
# Children과 Babies를 Kids로 통일해서 컬럼 생성
hotel['Kids'] = hotel.Children + hotel.Babies

# 예약별 총 투숙인원 컬럼 생성
hotel['total_members'] = hotel.Adults + hotel.Children + hotel.Babies
hotel
```

```python
# 모델 파이프라인 구성
rf_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                              ('regressor', RandomForestRegressor(random_state=42))])

xgb_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                               ('regressor', XGBRegressor(random_state=42))])

# 하이퍼파라미터 그리드 설정
rf_param_grid = {
    'regressor__n_estimators': [100, 200],
    'regressor__max_depth': [None, 10, 20],
}

xgb_param_grid = {
    'regressor__n_estimators': [100, 200],
    'regressor__max_depth': [3, 5, 7],
}

# 그리드 서치를 통한 모델 최적화
rf_grid_search = GridSearchCV(rf_pipeline, rf_param_grid, cv=5, scoring='neg_mean_squared_error')
rf_grid_search.fit(X_train, y_train)

xgb_grid_search = GridSearchCV(xgb_pipeline, xgb_param_grid, cv=5, scoring='neg_mean_squared_error')
xgb_grid_search.fit(X_train, y_train)
```

```python
# 최적의 랜덤 포레스트 모델 평가
rf_best_model = rf_grid_search.best_estimator_
rf_pred = rf_best_model.predict(X_test)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
print(f'랜덤 포레스트 최적 모델 RMSE: {rf_rmse}')

# 최적의 XGBoost 모델 평가
xgb_best_model = xgb_grid_search.best_estimator_
xgb_pred = xgb_best_model.predict(X_test)
xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_pred))
print(f'XGBoost 최적 모델 RMSE: {xgb_rmse}')
```
